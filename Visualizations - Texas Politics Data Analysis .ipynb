```python
#import packages
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.simplefilter("ignore")
```

```python
#import dataframe created in previous analysis 
data_2015 = pd.read_csv('texas_data.csv')
data_2011 = pd.read_csv('texas_data_2011.csv')
data_2018 = pd.read_csv('demo_2018.csv')

```

## Percentage of Hispanics in Texas counties in 2011, 2015, and 2018

```python
x= data_2011['hispanic_percent']
y= data_2015['hispanic_percent']
z= data_2018['hispanic_percent']

f, axes = plt.subplots(1,3, figsize=(20,10), sharex=True)
sns.despine(left=True)
sns.distplot(x, label='2011', ax=axes[0]).set_title("2011")
sns.distplot(y,label='2015', ax=axes[1]).set_title("2015")
sns.distplot(z, label='2018', ax=axes[2]).set_title("2018")
```

```python
print('Average % of hispanics in Texas counties in 2011: %',round(np.mean(x)*100,2))
print('Average % of hispanics in Texas counties in 2015: %',round(np.mean(y)*100,2))
print('Average % of hispanics in Texas counties in 2018: %',round(np.mean(z)*100,2))
```

## Variable relationships and correlations

```python
sns.lmplot(x='romney_percent', y='trump_percent', data=data_2015)
```

```python
#visualize correlation matrix

import seaborn as sns
# Compute the correlation matrix
corr = data_2015.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(10, 8))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
#pair plot of all variables 
sns.pairplot(data_2015)
```

```python
#plotting relationship between hispanic population and white population in Texas counties 
sns.jointplot(x="hispanic_percent", y="white_percent", data=data_2015, kind="kde")
```

```python
#relationship between % of white people in a county and % voting for trump
sns.jointplot(x="white_percent", y="trump_percent", data=data_2015, kind="kde")
```

```python
#relationship between % of white people and % of romney voters
sns.jointplot(x="white_percent", y="romney_percent", data=data_2015, kind="kde")
```

## 2015 Distribution of Race in Texas Counties

```python
f, axes = plt.subplots(2, 2, figsize=(20,10), sharex=True)
sns.despine(left=True)
sns.boxplot(x='white_percent', data=data_2015, palette = 'bright', ax=axes[0,0])
sns.boxplot(x='black_percent', data=data_2015, palette = 'bright', ax=axes[0,1])
sns.boxplot(x='hispanic_percent', data=data_2015, palette = 'bright', ax=axes[1,0])
sns.boxplot(x='other_race_percent', data=data_2015, palette = 'bright', ax=axes[1,1])
```

## 2018 Distribution of Race in Texas Counties

```python
f, axes = plt.subplots(2, 2, figsize=(20,10), sharex=True)
sns.despine(left=True)
sns.boxplot(x='white_percent', data=data_2018, palette = 'pastel', ax=axes[0,0])
sns.boxplot(x='black_percent', data=data_2018, palette = 'pastel', ax=axes[0,1])
sns.boxplot(x='hispanic_percent', data=data_2018, palette = 'pastel', ax=axes[1,0])
sns.boxplot(x='other_race_percent', data=data_2018, palette = 'pastel', ax=axes[1,1])
```
